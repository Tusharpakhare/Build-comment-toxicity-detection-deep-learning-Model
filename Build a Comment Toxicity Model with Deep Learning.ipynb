# preprocessing

```python
#tokenaization useing keras 
from tensorflow.keras.layers import TextVectorization
```

```python
X = df['comment_text']
y= df[df.columns[2:]].values
```

```python
MAX_FEATURES=20000 #num of word in vocab
```

```python
vectorizer = TextVectorization(max_tokens=MAX_FEATURES,
                              output_sequence_length=1800,
                              output_mode='int')
```

```python
vectorizer.adapt(X.values)#
```

```python
vectorizer.get_vocabulary()
```

```python
vectorized_text=vectorizer(X.values)#convert into array
```

```python
vectorized_text
```

```python
#tensorflow dataset  pipeline
#map,cache,shuffle,batch,prefetch from tensorflow_sclice,list_file
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8)  #help to prevent bottlenecks
```

```python
batch_X,batch_y=dataset.as_numpy_iterator().next()
```

```python
train=dataset.take(int(len(dataset)*.7))
val=dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test=dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))                
```

# Create Sequential model

```python
#1.embeding layer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional,Dense
from keras.layers import Embedding
```

```python
model = Sequential()
#create a embeding layers
model.add(Embedding(MAX_FEATURES+1,32))
#create a LSTM layers
model.add(Bidirectional(LSTM(32,activation='tanh')))#bidirectional-add info to both direction
#feature extractor fully connected layer
model.add(Dense(128,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
#final layer
model.add(Dense(6,activation='sigmoid'))
```

```python
model.compile(loss='BinaryCrossentropy',optimizer='Adam')
```

```python
model.summary()
```

```python
history=model.fit(train,epochs=1,validation_data=val) 
```

# Evaluating Model

```python
from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    # Unpack the batch 
    X_true, y_true = batch
    # Make a prediction 
    yhat = model.predict(X_true)
    
    # Flatten the predictions
    y_true = y_true.flatten()
    yhat = yhat.flatten()
    
    pre.update_state(y_true, yhat)
    re.update_state(y_true, yhat)
    acc.update_state(y_true, yhat)
```

```python
print(f'Precision: {pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```

```python
def score_comment(comment):
    vectorized_comment = vectorizer([comment])
    results = model.predict(vectorized_comment)
    
    text = ''
    for idx, col in enumerate(df.columns[2:]):
        text += '{}: {}\n'.format(col, results[0][idx]>0.5)
    
    return text
```

```python
interface = gr.Interface(fn=score_comment, 
                         inputs=gr.inputs.Textbox(lines=2, placeholder='Comment to score'),
                        outputs='text')
```
